```python
url = 'https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=437'
page = urlopen(url)
soup = BeautifulSoup(page, 'html.parser')
res=[]
for j in soup.find(class_='type06_headline').find_all('li'):
    url = j.find('a')['href']
    soup = BeautifulSoup(urlopen(url), 'html.parser')
    tmp = [i for i in soup.find_all('a') if str(i.get_text).find('href')!=-1]
    url = [i['href'] for i in tmp if i['href'].find('.jtbc.co.kr/article/article')!=-1][0]
    soup = BeautifulSoup(urlopen(url), 'html.parser')
    if soup.find(id='jtbcBody').find(class_='artical_date').text.split()[1]=='2023-07-14':
        res.append({'제목': soup.find(id='jtbcBody').find(id='jtbcBody').text,
        '본문': soup.find(class_='article_main').find(class_='article_content').text,
        'url': url})
res
```

```python
# 다른 분의 코드
url = 'https://finance.naver.com/item/sise_day.naver?code=005930'
r = requests.get(url, headers = {'User-Agent':'Mozilla/5.0}'})
s = BeautifulSoup(r.text, 'html.parser')
num_max = int(s.select('a[href*="item/sise_day"]')[-1].attrs['href'][-3:])+1

start_time = time.time()
res_df = pd.DataFrame(columns = ['날짜', '종가', '전일비', '시가', '고가', '저가', '거래량'])
for page in range(1,num_max):
        r = requests.get(url+f'&page={page}', headers = {'User-Agent':'Mozilla/5.0}'})
        # s = BeautifulSoup(r.text, 'html.parser')
        tmp_df = pd.read_html(r.text)[0].dropna(axis=0, how='all')
        res_df = pd.concat([res_df, tmp_df])
print(res_df.reset_index(drop=True))
time.time() - start_time
```
