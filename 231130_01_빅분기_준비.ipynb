```python
display(x_test.head())
display(y_test.head())
```

```python
x_train['Gender'] = x_train['Gender'].apply(str.strip).apply(str.lower)
for _ in x_train.columns:
    print(_, x_train[_].dtype, x_train[_].nunique())
    if x_train[_].dtype==object and x_train[_].nunique()<10:
        print(x_train[_].unique())
# pd.get_dummies()
x_train['Female'] = pd.get_dummies(x_train['Gender'], dtype=int).female
x_train_modified = pd.concat([x_train, pd.get_dummies(x_train['Geography'], dtype=int)], axis=1).drop(columns=['CustomerId', 'Surname', 'Geography', 'Gender'])
x_train_modified
```

```python
x_test['Gender'] = x_test['Gender'].apply(str.strip).apply(str.lower)
for _ in x_test.columns:
    print(_, x_test[_].dtype, x_test[_].nunique())
    if x_test[_].dtype==object and x_test[_].nunique()<10:
        print(x_test[_].unique())
# pd.get_dummies()
x_test['Female'] = pd.get_dummies(x_test['Gender'], dtype=int).female
x_test_modified = pd.concat([x_test, pd.get_dummies(x_test['Geography'], dtype=int)], axis=1).drop(columns=['CustomerId', 'Surname', 'Geography', 'Gender'])
x_test_modified
```

```python
from scipy.stats import ttest_1samp
from scipy.stats import shapiro
from scipy.stats import wilcoxon

# 정규성 검정 샤피로
static, pvalue = shapiro(df)
print('샤피로 정규성 검정 p-value : ',pvalue,'이므로')

if pvalue < 0.05:
    print('귀무가설을 기각한다. 정규성을 만족하지 않으므로 비모수 검정을 진행한다. 윌콕슨 순위 부호 검정을 진행한다.\n')
    print('윌콕슨 순위 부호 검정의 귀무가설은 "100명 키의 평균은 165이다." 이며 대립가설은 "100명 키의 평균은 165가 아니다." 이다')
    
    #윌콕슨 부호순위 검정
    static, pvalue = wilcoxon(df['height']-165) # or static, pvalue = wilcoxon(df['height'], np.ones(len(df)) *165)
    
    
    
    if pvalue < 0.05:
        print(f'검정 결과 pvalue는 {pvalue}로 결과는 귀무가설을 기각하고 대립가설을 채택한다.')
    else:
        print(f'검정 결과 pvalue는 {pvalue}로 결과는 귀무가설을 기각하지 않는다.')
    
    
else:
    print('귀무가설을 기각하지 않는다. 정규성을 만족하므로 단일표본 검정으로 확인한다.\n')
    print('단일표본 t-test의 귀무가설은 "100명 키의 평균은 165이다." 이며 대립가설은 "100명 키의 평균은 165가 아니다." 이다')
    
    #단일 표본 t 검정
    static, pvalue = ttest_1samp(df['height'],165) 
    if pvalue < 0.05:
        print(f'검정 결과 pvalue는 {pvalue}로 결과는 귀무가설을 기각하고 대립가설을 채택한다.')
    else:
        print(f'검정 결과 pvalue는 {pvalue}로 결과는 귀무가설을 기각하지 않는다.')  
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# 남성과 여성의 인구 수
total_male = 600
total_female = 550

# 남성과 여성 중 흡연자의 비율
smoking_ratio_male = 0.2
smoking_ratio_female = 0.26

# 흡연자와 비흡연자의 인구 수 계산
smoking_male = int(total_male * smoking_ratio_male)
non_smoking_male = int(total_male - smoking_male)

smoking_female = total_female * smoking_ratio_female
non_smoking_female = total_female - smoking_female

# 데이터 배열 생성 (빈도로 변환)
data = np.array([[smoking_male, non_smoking_male], [smoking_female, non_smoking_female]])

# 카이제곱 검정 수행
chi2_stat, p_val, dof, expected = chi2_contingency(data)

print('기각',p_val)
```

```python
import pandas as pd 
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/adp/28/p7.csv')

import statsmodels.api as sm
df['inter'] = 1.
df = df[['age','Cholesterol','inter','weight']]
lm = sm.OLS(df.weight, df.drop(columns='weight')).fit()
print(lm.summary())
print(lm.params['age'])
print(lm.pvalues['Cholesterol'])
print(lm.predict([55,70.6,1]))
```

```python
import pandas as pd 
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/adp/28/p7.csv')

import statsmodels.api as sm
X = sm.add_constant(df[['age', 'Cholesterol']]) 
model = sm.OLS(df['weight'], X)
results = model.fit()

# 전체 회귀 결과 출력
print(results.summary())

print(results.params['age'])
print(results.pvalues['Cholesterol'])
pred = results.predict([1,55,72.6]) # const , age,Cholesterol
print(pred[0])
```
