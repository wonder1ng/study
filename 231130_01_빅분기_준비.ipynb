```python
sum(df.groupby(['channelTitle', 'channelId']).count().reset_index().groupby('channelId').count().channelTitle>1)
```

```python
sum(df[['channelTitle','channelId']].drop_duplicates().channelId.value_counts()>1)
```

```python
df[pd.to_datetime(df.trending_date2).dt.day_of_week==6].value_counts('categoryId').index[0]
```
